```python
#1. Introduction
#The Iris dataset is one of the most famous datasets in machine learning.
#It is used for classification problems.
#ðŸŽ¯ Objective:
#To analyze flower measurements and understand differences between species using EDA (Exploratory Data Analysis).
#Dataset contains 3 species:
#Setosa
#Versicolor
#Virginica
#Features:
#Feature	Description
#sepal_length	Length of sepal (cm)
#sepal_width	Width of sepal (cm)
#petal_length	Length of petal (cm)
#petal_width	Width of petal (cm)
#species	Flower category (Target)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.use('tkagg')
#Load Iris dataset
iris = sns.load_dataset("iris")
#InspectION
print(iris.head())
#Dataset Shape:
#150 rows
#5 columns
print("Shape of dataset:", iris.shape)
print("Columns:", iris.columns)
print(iris.info())
print(iris.describe())
#Scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
#Histogram
plt.figure(figsize=(8,5))
sns.histplot(iris["petal_length"], bins=20, kde=True)
plt.title("Distribution of Petal Length")
plt.show()
#Box plot
plt.figure(figsize=(8,5))
sns.boxplot(data=iris, x="species", y="petal_width")
plt.title("Petal Width by Species")
plt.show()
input("Press Enter to exit...")
#Accuracy is usually around 95% â€“ 100%
```
